```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np, pandas as pd
from scipy.stats import bernoulli, binom, poisson, gamma, weibull_min
import seaborn as sns
```

# Section 1a

```python
probability = 0.07
def exptVal(p):
    return bernoulli.mean(p)
print('Expected value is: ', exptVal(probability))
```

```python
print(f"Vote For is {round(probability*100)}%")
print(f"Vote Against is {(1-probability)*100}%")
```

```python
bernoulli_dist = bernoulli.rvs(p=0.07,size=100)
sns.distplot(bernoulli_dist, kde=False)
plt.title('Bernoulli Distribution')
plt.xlabel('0 - Vote Against & 1 - Vote For')
plt.ylabel('Proportion')
plt.show()
```

# Section 1b

```python
ep=0.0
k=1
mu = 44
while ep < 0.005:
    ep = poisson.cdf(k,mu)
    k+=1

#Plotting the graph
fig, ax = plt.subplots(1, 1)
x = np.arange(1,k-1)

ax.plot(x, np.round(poisson.cdf(x, mu),3), 'bo', ms=8, label='poisson cdf')
ax.vlines(x, 0, np.round(poisson.cdf(x, mu),3), colors='b', lw=5, alpha=0.5)


ax.set_title('Meteorites falling with their probabilities')
plt.xlabel('Number of Meteorites falling in a year',fontsize=14)
plt.ylabel('Probability',fontsize=14)

plt.show()
```

```python
data_binom = poisson.rvs(mu=44, size=100)
ax = sns.distplot(data_binom,
                  kde=True,
                  color='skyblue',
                  hist_kws={"linewidth": 5,'alpha':1})
plt.axvline(x=poisson.median(mu, loc=0),
            color='red')
plt.xlabel('Poisson distribution of sample')
plt.ylabel('Frequency')
plt.title('Median and Variance')

median= poisson.median(mu, loc=0)
var = poisson.var(mu, loc=0)

median_patch = mpatches.Patch(color='blue', label=f'Median: {median}')
var_patch = mpatches.Patch(color='y', label=f'Variance: {var}')
plt.legend(handles=[median_patch,var_patch])

plt.show()
```

```python

hour2= (0.30*np.exp(-0.5*2)) + (0.69*np.exp(-0.25*2))
hour4 = (0.30*np.exp(-0.5*4)) + (0.69*np.exp(-0.25*4))
Probability2 = 1-hour2
Probability4 = 1-hour4
print(Probability4-Probability2)
```

```python
y=np.arange(2,4,0.0192)
minimum=[]
Probability=[]
for i in np.round(y,2):
    p = (0.30*np.exp(-0.5*i)) + (0.69*np.exp(-0.25*i))
    minimum.append(int(np.round(i*60 , 0)))
    Probability.append(np.round(p,3))

dataframe = pd.DataFrame({'Minutes':minimum, 'Probability':Probability})
dataframe.head()
```

```python
plt.plot(dataframe.Minutes,dataframe.Probability)
plt.ylabel('Probability')
plt.xlabel('Owl wait time in minutes from 2 to 4 hours')
plt.show()
```
